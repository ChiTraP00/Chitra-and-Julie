### UK

```python
sub_loop_uktype = UK_Total_updated[['A', 'B', 'C10_C12', 'C13_C15', 'C16', 'C17_C18', 'C19', 'C20_C22', 'C23', 'C24_C25', 'C26_C30', 'C31_C33', 'D', 'E36_E37_E39', 'F', 'G_U_X_G4677', 'EP_HH']]
for col in sub_loop_uktype:
    cross = pd.crosstab(UK_Total_updated['Waste_types'], UK_Total_updated[col])
    results = stats.chi2_contingency(cross)
    print(col, ":", "P-value:",results[1], "Assumption passed:", results[3]) # > 5
```

```python
uk_loc = UK_Total_updated.drop(['Unnamed: 0', 'Haz_NonHaz'], axis=1)
```

```python
x = uk_loc.iloc[:,:20].drop(["Year", "Waste_types"], axis=1)
y = uk_loc["Waste_types"] 
```

```python
# what model you want to use... and then cross validate it.
# We choose the classifier 
```

```python
rf = RandomForestClassifier()
```

```python
# Train/ Test / Split
# Intial Random Forest Model
# Evaluate Model Fit 

# By running cross_val_score() function, you can skip all the steps above
# go straight to cross validation. 
```

```python
cross_val_score(rf, x , y, cv=3)
```

```python
## Accurcy score of  model as a whole before splitting it for training and testing 
```

```python
# cross valdation testing elimates the biast data and shows the best percentage array.

# (helps elimate biast, by using default values, they take a section of the dataset, 
# and creates a model, test data, and the rest is training data. open data
# training data will be the data that if fresh and nw, different data
# will be tested and trained and new. 
# the array, is the 1st model got 90% etc. and then take the average of all the models
# 1/3 iteration) 
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3, random_state=76)
n_estimators_array = [1, 4, 5, 8, 10, 20, 50, 75, 100, 250, 500]
results = []
for n in n_estimators_array:
    forest = RandomForestClassifier(n_estimators=n, random_state=76)
    forest.fit(x_train, y_train)
    result = accuracy_score(y_test, forest.predict(x_test))
    results.append(result) 
    print(n, ':', result)
```

```python
# invididaul hyper perameter. (this can pin point just one perameter.)
# estimators
# this is how many desision trees your performing and which ones are the best
# deminshing returns? increase of accurancy, not worth the time
# 75 realistic and practice( takes alot of time)
```

```python
plt.plot(n_estimators_array, results)
```

```python
# plots the of accurrnacy of the values, you could put in estimantors if you wanted too.
# when looking at the graph you can see that around 250, the graph increases
# which means the accurancy increases
# safe to say 250 is the best point
```

```python
from sklearn.model_selection import RandomizedSearchCV
# Number of features to consider at every split
max_features = ['auto', None, 'log2']
# Maximum number of levels in tree
max_depth = [10, 20, 30, 40, 50, 60, 70, 80, 90, None]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 4]
# Method of selecting samples for training each tree
random_grid = {'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_leaf': min_samples_leaf}
print(random_grid)
```

```python
# max features: how many features will use in the tree, (it will potentially select them all)...
# max depth: nodes are in the tree (google it)... 
# min samples: ...
# dictionary: variables you want to experiment with...
```

```python
rf = RandomForestClassifier(n_estimators=75)
# Random search of parameters, using 3 fold cross validation, 
# search across 100 different combinations, and use all available cores
rf_random = RandomizedSearchCV(estimator = rf, param_distributions = random_grid, n_iter = 90, cv = 3, random_state=42)
# Fit the random search model
rf_random.fit(x_train, y_train)
```

```python
# this takes awhile, 
# rf random search ties every combination of values that you specify.
# for every iteration it does cross validation
# goes through all the combinations and say which one performs the best
```

```python
rf_random.best_params_
```
