## Data Wrangle

#### Spliting Total and Haz/Non-Haz into two different dataframes of Industries Waste Gen

loc -> location 

calling the dataset.loc (in that df we only want where H/NH split == and != Total)

```python
england_industry_total = england_industry_df.loc[england_industry_df['Hazardous/Non-hazardous split'] == "Total"]
england_industry_type = england_industry_df.loc[england_industry_df['Hazardous/Non-hazardous split'] != "Total"]
```

#### Spliting Total and Haz/Non-Haz into two different dataframes of Management Waste Treated 

```python
england_managment_total = england_managment_df.loc[england_managment_df['Hazardous/Non-hazardous split'] == "Total"]
england_managment_type = england_managment_df.loc[england_managment_df['Hazardous/Non-hazardous split'] != "Total"]
```

### *England Total -> Indusrty & Management* 

```python
england_managment_total
```

#### Combining two subject indentifier -> Year and EWC-STAT description -> which will make each row unique

- They did not match on the rows that were aggregates "Total waste generated" and "Total waste managed"

- So we removed those because they were rows that represented total's

```python
combine_yr_ewc = england_industry_total[['Year', 'EWC-STAT description']] == england_managment_total[['Year', 'EWC-STAT description']]
```

```python
combine_yr_ewc
```

Results: the years match but the EWC-STAT do not (101, 203, 305, 407)

#### loc-> locating

- calling all the False (i.e. all the total waste from different years- ) from the column EWC to see aren't the same, so we can remove before we concatinate meaning combine two worksheets.

- if == True: it would return all rows where they do match, which would be all except 101, 203, 305, 407
- if == False: it would return all rows where they do not match, which would be 101, 203, 305, 407

```python
combine_yr_ewc.loc[combine_yr_ewc['EWC-STAT description'] == False]
```

#### Printing the length before we remove 'Total waste generation'

```python
print(len(england_industry_total))
len(england_managment_total)
```

#### Row Manipulation: Calling all the rows to remove the rows that have 'Total waste generation'

- Double Checking to make sure: Printing the length after remove 'Total waste generation'

```python
england_industry_total = england_industry_total.loc[england_industry_total['EWC-STAT description'] != '*Total waste generation',:]
len(england_industry_total)
```

#### Row Manipulation: Calling all the rows to remove the rows that have 'Total waste treated'

- Double Checking to make sure: Printing the length after remove 'Total waste treated'

```python
england_managment_total = england_managment_total.loc[england_managment_total['EWC-STAT description'] != '*Total waste treated',:]
len(england_managment_total)
```

#### Merge Worksheets

- Above lines, we made sure Year and EWC Swat were exact same before we concatenated datasets

- Now that both tables are set to True Bouleans -> meaning removed total waste gen and treated; can combine both worksheets now that the unique subject identifier are both aligned

```python
total_full = pd.concat([england_industry_total, england_managment_total], axis=1)
```

```python
len(total_full)
```

```python
total_full
```

#### *England Haz/Non-Haz -> Indusrty & Management* 

```python
combine_yr_ewc2 = england_industry_type[['Year', 'EWC-STAT description']] == england_managment_type[['Year', 'EWC-STAT description']]
combine_yr_ewc2
```

```python
combine_yr_ewc2.loc[combine_yr_ewc2['EWC-STAT description'] == False]
```
