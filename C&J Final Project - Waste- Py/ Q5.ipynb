```python
def recode (series):
    if series == "Spent solvent":
        return 1
    if series == "Acid, alkaline or saline wastes":
        return 2
    if series == "Used oils":
        return 3
    if series == "Chemical wastes":
        return 4
    if series == "Industrial effluent sludges":
        return 5
    if series == "Sludges & liquid wastes from waste treatment":
        return 6
    if series == "Health care & biological wastes":
        return 7
    if series == "Metallic wastes, ferrous":
        return 8
    if series == "Metallic wastes, non-ferrous":
        return 9
    if series == "Metallic wastes, mixed":
        return 10
    if series == "Glass wastes":
        return 11
    if series == "Paper & cardboard wastes":
        return 12
    if series == "Rubber wastes":
        return 13
    if series == "Plastic wastes":
        return 14
    if series == "Wood wastes":
        return 15
    if series == "Textile wastes":
        return 16
    if series == "Waste containing PCB":
        return 17
    if series == "Discarded equipment":
        return 18
    if series == "Discarded vehicles":
        return 19
    if series == "Batteries & accumulators wastes":
        return 20
    if series == "Animal & mixed food waste":
        return 21
    if series == "Vegetal wastes":
        return 22
    if series == "Animal faeces, urine & manure":
        return 23
    if series == "Household & similar wastes":
        return 24
    if series == "Mixed & undifferentiated materials":
        return 25
    if series == "Sorting residues":
        return 25
    if series == "Common sludges":
        return 27
    if series == "Mineral waste from construction & demolition":
        return 28
    if series == "Other mineral wastes":
        return 29
    if series == "Combustion wastes":
        return 30
    if series == "Soils":
        return 31
    if series == "Dredging spoils":
        return 32
    if series == "Mineral waste from waste treatment & stabilised waste":
        return 33
    
uk_treat_t['Waste_typesR'] = uk_treat_t['Waste_types'].apply(recode)
```
